```python
movies = movies[["movie_id","title","cast","crew","genres","overview","keywords"]]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(obj):
    names = []
    for i in ast.literal_eval(obj):
        names.append(i["name"])
    return names
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head(2)
```

```python
def convert2(obj):
    names = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            names.append(i["name"])
            counter+=1
        else:
            break
    return names
```

```python
movies["cast"] = movies["cast"].apply(convert2)
```

```python
movies.head(2)
```

```python
def fetch_director(obj):
    names = []
    for x in ast.literal_eval(obj):
        if x['job'] == "Director":
            names.append(x['name'])
            break
    return names
```

```python
movies["crew"].values #"job": "Director"
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies["overview"][0]
```

```python
movies["overview"]=movies["overview"].apply(lambda x: x.split())
```

```python
movies['genres']= movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['genres'] + movies['cast'] + movies['crew'] + movies['overview'] + movies['keywords']
```

```python
movies_new_df = movies[["movie_id","title","tags"]]
```

```python
movies_new_df["tags"]=movies_new_df["tags"].apply(lambda x: " ".join(x))
```

```python
movies_new_df.head(2)
```

```python
movies_new_df['tags'] = movies_new_df['tags'].apply(lambda x: x.lower())
```

```python
movies_new_df.head(2)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer #loved loving love = love love love changes common words into one
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_new_df['tags'] = movies_new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words = "english") #convert into vectors and stop english common words like in, at, on, etc.
```

```python
vectors = cv.fit_transform(movies_new_df["tags"]).toarray()
```

```python
cv.get_feature_names()
```

```python
#we will caculate cosine distance instead of uniclined distance as tip to tip distance...
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movies_index = movies_new_df[movies_new_df['title']==movie].index[0]
    distances = similar[movies_index]
    movies_closed = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]
    
    for i in movies_closed:
        print(movies_new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 3')
```
